```python
# Requiere: qutip, numpy, matplotlib
import numpy as np
import qutip as qt
import matplotlib.pyplot as plt

# ---------------------------
# Parmetros (adimensionales, _b = 1)
# ---------------------------
omega_b = 1.0
lam_over_ob   = 0.03     # /_b (dbil p/ Fig. 3A)
Omega_over_ob = 0.003    # /_b (bombeo dbil)
kappa_over_ob = 0.002    # /_b
gamma_over_ob = 0.0002   # /_b
gphi_over_ob  = 0.0004   # _/_b

# Rango de /_b: de 0 a -5 (descendente, como pediste)
Delta_list = np.linspace(0.0, -5.0, 501)

# Truncacin del oscilador (suficiente para ver n=2 con prdidas pequeas)
Ncut = 14

# ---------------------------
# Operadores
# ---------------------------
b   = qt.destroy(Ncut)
I_b = qt.qeye(Ncut)
sm  = qt.sigmam()
sp  = sm.dag()
I_q = qt.qeye(2)
proj_e = qt.tensor(sp*sm, I_b)        # 
b_sys  = qt.tensor(I_q, b)
sm_sys = qt.tensor(sm, I_b)
sp_sys = sm_sys.dag()

# ---------------------------
# Funcin para g^(2)(0)
# g2 = <b b b b> / <b b>^2
# ---------------------------
def g2_of_rho(rho):
    nbar = qt.expect(b_sys.dag()*b_sys, rho)
    if nbar <= 1e-14:
        return np.nan
    num = qt.expect(b_sys.dag()*b_sys.dag()*b_sys*b_sys, rho)
    return np.real_if_close(num/(nbar**2))

# ---------------------------
# Barrido en  y clculo de estado estacionario
# ---------------------------
g2_vals = np.full_like(Delta_list, np.nan, dtype=float)

# Colapsos (tres canales: b, , )
c_ops = [
    np.sqrt(kappa_over_ob) * b_sys,
    np.sqrt(gamma_over_ob) * sm_sys,
    np.sqrt(gphi_over_ob)  * proj_e
]

# Trminos estticos que no dependen de 
H_bare  = omega_b * qt.tensor(I_q, b.dag()*b)
H_int   = lam_over_ob * qt.tensor(sp*sm, (b + b.dag()))
H_drive = Omega_over_ob * (sm_sys + sp_sys)

for i, Delta_over_ob in enumerate(Delta_list):
    H = (Delta_over_ob * proj_e) + H_bare + H_int + H_drive
    rho_ss = qt.steadystate(H, c_ops, method='direct')
    g2_vals[i] = g2_of_rho(rho_ss)

# ---------------------------
# Grfica: g^(2) vs /_b
# ---------------------------
fig, ax = plt.subplots(figsize=(7.5, 4.2))
ax.plot(Delta_list, g2_vals, lw=1.5)

ax.set_yscale('log')
ax.set_ylim(0, 1e10)
ax.set_xlim(0.0, -5.0)   
ax.set_xlabel(r'$\Delta/\omega_b$')
ax.set_ylabel(r'$g^{(2)}(0)$')

ax.set_title(r'$g^{(2)}(0)$ vs $\Delta/\omega_b$')
ax.grid(True, which='both', ls=':', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# Mapa tipo Fig. 3(b) SIN curvas _n
# X: /_b (0  -6), Y: /_b (0  10^0),
# Colorbar: 0..9 = log10(g^{(2)}(0)) (verde=0, amarillo=9)

import numpy as np
import qutip as qt
import matplotlib.pyplot as plt

# ---------------------------
# Parmetros (_b = 1)
# ---------------------------
omega_b = 1.0
Omega_over_ob = 0.003     # /_b (bombeo dbil)
kappa_over_ob = 0.002     # /_b
gamma_over_ob = 0.0002    # /_b
gphi_over_ob  = 0.0004    # _/_b

# Barridos
Delta_list  = np.linspace(0.0, -6.0, 361)   # eje X: 0  -6
lambda_list = np.linspace(0.00, 1.00, 81)   # eje Y: 0  1 (=10^0)

# Truncacin del oscilador
Ncut = 18

# ---------------------------
# Operadores
# ---------------------------
b   = qt.destroy(Ncut)
I_b = qt.qeye(Ncut)
sm  = qt.sigmam()
sp  = sm.dag()
I_q = qt.qeye(2)

proj_e = qt.tensor(sp*sm, I_b)        # 
b_sys  = qt.tensor(I_q, b)
sm_sys = qt.tensor(sm, I_b)
sp_sys = sm_sys.dag()
nb_sys = b_sys.dag()*b_sys

# ---------------------------
# Colapsos (Lindblad)
# ---------------------------
c_ops = [
    np.sqrt(kappa_over_ob) * b_sys,   #  D[b]
    np.sqrt(gamma_over_ob) * sm_sys,  #  D[]
    np.sqrt(gphi_over_ob)  * proj_e   #  D[]
]

# Partes de H que no dependen de  ni 
H_bare  = omega_b * qt.tensor(I_q, b.dag()*b)          # _b bb
H_drive = Omega_over_ob * (sm_sys + sp_sys)            # (+)

# ---------------------------
# g^(2)(0) con piso de ocupacin
# ---------------------------
def g2_of_rho(rho, nbar_floor=1e-10):
    nbar = qt.expect(nb_sys, rho)
    if nbar <= nbar_floor:
        return np.nan
    num = qt.expect(b_sys.dag()*b_sys.dag()*b_sys*b_sys, rho)
    return float(np.real(num / (nbar**2)))

# ---------------------------
# Mapa g2(, )
# ---------------------------
g2_map = np.full((len(lambda_list), len(Delta_list)), np.nan, dtype=float)

for iL, lam_over_ob in enumerate(lambda_list):
    H_int = lam_over_ob * qt.tensor(sp*sm, (b + b.dag()))  #   (b+b)
    for iD, Delta_over_ob in enumerate(Delta_list):
        H = (Delta_over_ob * proj_e) + H_bare + H_int + H_drive
        rho_ss = qt.steadystate(H, c_ops, method='direct')
        g2_map[iL, iD] = g2_of_rho(rho_ss, nbar_floor=1e-10)

# ---------------------------
# Transformacin a log10 y ploteo (0..9)
# ---------------------------
# log_g2 = log10(max(g2, 1))  -> todo g2<1 aparece como 0 (verde)
log_g2 = np.log10(np.maximum(g2_map, 1.0))
log_g2 = np.clip(log_g2, 0.0, 9.0)

fig, ax = plt.subplots(figsize=(8.8, 5.0))
im = ax.imshow(
    log_g2,
    aspect='auto',
    origin='lower',
    extent=[Delta_list[0], Delta_list[-1], lambda_list[0], lambda_list[-1]],
    vmin=0.0, vmax=9.0,
    cmap='summer'  # verde -> amarillo
)

cbar = plt.colorbar(im, ax=ax, pad=0.02)
cbar.set_label(r'$\log_{10}\, g^{(2)}(0)$')
cbar.set_ticks(range(0,10))
cbar.set_ticklabels([str(t) for t in range(0,10)])

ax.set_xlabel(r'$\Delta/\omega_b$')       # X: 0  6
ax.set_ylabel(r'$\lambda/\omega_b$')      # Y: 0  1
ax.set_title(r'Mapa $\log_{10} g^{(2)}(0)$ (verde=0, amarillo=9)')
ax.set_xlim(0.0, -6.0)
ax.set_ylim(0.0, 1.0)
ax.grid(False)
plt.tight_layout()
plt.show()
```
