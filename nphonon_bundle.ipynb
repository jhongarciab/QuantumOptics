```python
import numpy as np
import qutip as qt
import matplotlib.pyplot as plt

omega_b = 1.0
lam_over_ob   = 0.03     
Omega_over_ob = 0.003    
kappa_over_ob = 0.002    
gamma_over_ob = 0.0002   
gphi_over_ob  = 0.0004   

Delta_list = np.linspace(0.0, -5.0, 501)

Ncut = 14

b   = qt.destroy(Ncut)
I_b = qt.qeye(Ncut)
sm  = qt.sigmam()
sp  = sm.dag()
I_q = qt.qeye(2)
proj_e = qt.tensor(sp*sm, I_b)      
b_sys  = qt.tensor(I_q, b)
sm_sys = qt.tensor(sm, I_b)
sp_sys = sm_sys.dag()

def g2_of_rho(rho):
    nbar = qt.expect(b_sys.dag()*b_sys, rho)
    if nbar <= 1e-14:
        return np.nan
    num = qt.expect(b_sys.dag()*b_sys.dag()*b_sys*b_sys, rho)
    return np.real_if_close(num/(nbar**2))

g2_vals = np.full_like(Delta_list, np.nan, dtype=float)

c_ops = [
    np.sqrt(kappa_over_ob) * b_sys,
    np.sqrt(gamma_over_ob) * sm_sys,
    np.sqrt(gphi_over_ob)  * proj_e
]

H_bare  = omega_b * qt.tensor(I_q, b.dag()*b)
H_int   = lam_over_ob * qt.tensor(sp*sm, (b + b.dag()))
H_drive = Omega_over_ob * (sm_sys + sp_sys)

for i, Delta_over_ob in enumerate(Delta_list):
    H = (Delta_over_ob * proj_e) + H_bare + H_int + H_drive
    rho_ss = qt.steadystate(H, c_ops, method='direct')
    g2_vals[i] = g2_of_rho(rho_ss)

fig, ax = plt.subplots(figsize=(7.5, 4.2))
ax.plot(Delta_list, g2_vals, lw=1.5)

ax.set_yscale('log')
ax.set_ylim(0, 1e10)
ax.set_xlim(0.0, -5.0)   
ax.set_xlabel(r'$\Delta/\omega_b$')
ax.set_ylabel(r'$g^{(2)}(0)$')

ax.set_title(r'$g^{(2)}(0)$ vs $\Delta/\omega_b$')
ax.grid(True, which='both', ls=':', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
omegab = 1.0
n = 2                      
lambda_omegab = 0.1      
Omega_omegab  = 0.2      
Delta = -n * omegab        

kappa = 0.002               
gamma = 0.0002               
gamma_phi = 0.0004           
nth = 0.0                  

Nph = 12

b  = destroy(Nph)
nb = b.dag()*b
Ib = qeye(Nph)

sm = sigmam()
sp = sigmap()
Pv = ket2dm(basis(2,0))    
Pc = ket2dm(basis(2,1))    
Is = qeye(2)

b0 = basis(Nph, 0)
P0_b = ket2dm(b0)
P0v_op = tensor(P0_b, Pv)
P0c_op = tensor(P0_b, Pc)

lam = lambda_omegab * omegab  
Om  = Omega_omegab  * omegab  

H = (omegab * tensor(nb, Is)
     + Delta * tensor(Ib, Pc)
     + lam   * tensor(b + b.dag(), Pc)
     + 0.5*Om* tensor(Ib, sp + sm))

c_ops = []
if kappa > 0:
    if nth > 0:
        c_ops += [np.sqrt(kappa*(nth+1)) * tensor(b, Is)]
        c_ops += [np.sqrt(kappa*nth)     * tensor(b.dag(), Is)]
    else:
        c_ops += [np.sqrt(kappa)         * tensor(b, Is)]
if gamma > 0:
    c_ops += [np.sqrt(gamma) * tensor(Ib, sm)]
if gamma_phi > 0:
    c_ops += [np.sqrt(gamma_phi) * tensor(Ib, sigmaz()/2)]

psi0 = tensor(basis(Nph,0), basis(2,0))

x_min, x_max = 0.0, 4000.0
t_end = x_max / omegab              
Nt = 20001                           
tlist = np.linspace(0.0, t_end, Nt)

np.random.seed(1234)                
e_ops = [P0v_op, P0c_op]
res = mcsolve(H, psi0, tlist, c_ops, e_ops=e_ops, ntraj=1, progress_bar=None)

P0v_t = res.expect[0]   
P0c_t = res.expect[1]   

rcParams.update({"font.size": 11})
x = omegab * tlist                  

plt.figure(figsize=(9.0, 2.6))
plt.plot(x, P0v_t, color='red', lw=1.6, label=r'$P_{0v}$')
plt.plot(x, P0c_t, color='black',   lw=1.6, label=r'$P_{0c}$')
plt.xlabel(r'$\omega_b t$')
plt.ylabel('Probabilidad')

plt.xlim(x_min, x_max)
plt.ylim(0.0, 1.0)

plt.legend(loc='upper right', frameon=False)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import qutip as qt
import matplotlib.pyplot as plt
from qutip.visualization import matrix_histogram
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable

Delta_over_ob = -2.0     
lam_over_ob   = 0.1     
Omega_over_ob = 0.2     
kappa_over_ob = 0.002    
gamma_over_ob = 0.0002    
gphi_over_ob  = 0.0004    

N  = 3
a  = qt.tensor(qt.destroy(N), qt.qeye(2))   
sm = qt.tensor(qt.qeye(N), qt.sigmam())     
sp = sm.dag()
sx = qt.tensor(qt.qeye(N), qt.sigmax())      
proj_e = sp * sm                             

H = (
    Delta_over_ob * proj_e
    + (a.dag() * a)
    + lam_over_ob * proj_e * (a + a.dag())
    + Omega_over_ob * sx
)

c_ops = [
    np.sqrt(kappa_over_ob) * a,
    np.sqrt(gamma_over_ob) * sm,
    np.sqrt(gphi_over_ob)  * proj_e
]

ket_v, ket_c = qt.basis(2, 0), qt.basis(2, 1)
psi0 = (qt.tensor(qt.basis(N, 0), ket_v) + qt.tensor(qt.basis(N, 2), ket_c)).unit()

tmax  = 8.0 / max(kappa_over_ob, 1e-9)
tlist = np.linspace(0.0, tmax, 2001)

nb_sys = a.dag() * a
res_flux = qt.mesolve(H, psi0, tlist, c_ops, [nb_sys], progress_bar=None)
flux_t = kappa_over_ob * np.real(res_flux.expect[0])

idx_star = int(np.nanargmax(flux_t))
t_star   = float(tlist[idx_star])

rho_tstar = qt.mesolve(H, psi0, [t_star], c_ops, [], progress_bar=None).states[-1]

USE_SWAP = False 

def permutation_operator_from_old_indices(old_indices):
    """Dado el mapeo new_index -> old_index, crea P tal que |e_new> = P |e_old>."""
    dim = len(old_indices)
    P_mat = np.zeros((dim, dim), dtype=complex)
    for new_i, old_i in enumerate(old_indices):
        P_mat[new_i, old_i] = 1.0
    return qt.Qobj(P_mat, dims=[[2, N], [2, N]])

if USE_SWAP:
    old_indices = [1, 0, 3, 2, 5, 4]
    P = permutation_operator_from_old_indices(old_indices)
    rho_plot = P * rho_tstar * P.dag()
else:
    rho_plot = rho_tstar

kets_labels = [r"$|0,v\rangle$", r"$|0,c\rangle$",
               r"$|1,v\rangle$", r"$|1,c\rangle$",
               r"$|2,v\rangle$", r"$|2,c\rangle$"]

try:
    fig, ax = matrix_histogram(rho_plot, kets_labels, kets_labels,
                               limits=[0, 1], colorbar=False)
    add_manual_cbar = True
except TypeError:
    fig, ax = matrix_histogram(rho_plot, kets_labels, kets_labels, limits=[0, 1])
    add_manual_cbar = False

ax.view_init(azim=-55, elev=45)
ax.set_zlim(0, 1)

if add_manual_cbar:
    norm = Normalize(vmin=0.0, vmax=1.0)
    sm   = ScalarMappable(norm=norm, cmap=plt.get_cmap())
    sm.set_array([])
    cbar = plt.colorbar(sm, ax=ax, pad=0.06, shrink=0.7)
    cbar.set_ticks([0, 0.25, 0.5, 0.75, 1.0])
    cbar.set_label(r"$|\rho_{ij}|$")

plt.tight_layout()
plt.show()
```
