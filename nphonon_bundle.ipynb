```python
omega_b = 1.0
lam_over_ob   = 0.03     
Omega_over_ob = 0.003    
kappa_over_ob = 0.002    
gamma_over_ob = 0.0002   
gphi_over_ob  = 0.0004   

Delta_list = np.linspace(0.0, -5.0, 501)

Ncut = 14

b   = qt.destroy(Ncut)
I_b = qt.qeye(Ncut)
sm  = qt.sigmam()
sp  = sm.dag()
I_q = qt.qeye(2)
proj_e = qt.tensor(sp*sm, I_b)      
b_sys  = qt.tensor(I_q, b)
sm_sys = qt.tensor(sm, I_b)
sp_sys = sm_sys.dag()

def g2_of_rho(rho):
    nbar = qt.expect(b_sys.dag()*b_sys, rho)
    if nbar <= 1e-14:
        return np.nan
    num = qt.expect(b_sys.dag()*b_sys.dag()*b_sys*b_sys, rho)
    return np.real_if_close(num/(nbar**2))

g2_vals = np.full_like(Delta_list, np.nan, dtype=float)

c_ops = [
    np.sqrt(kappa_over_ob) * b_sys,
    np.sqrt(gamma_over_ob) * sm_sys,
    np.sqrt(gphi_over_ob)  * proj_e
]

H_bare  = omega_b * qt.tensor(I_q, b.dag()*b)
H_int   = lam_over_ob * qt.tensor(sp*sm, (b + b.dag()))
H_drive = Omega_over_ob * (sm_sys + sp_sys)

for i, Delta_over_ob in enumerate(Delta_list):
    H = (Delta_over_ob * proj_e) + H_bare + H_int + H_drive
    rho_ss = qt.steadystate(H, c_ops, method='direct')
    g2_vals[i] = g2_of_rho(rho_ss)

fig, ax = plt.subplots(figsize=(7.5, 4.2))
ax.plot(Delta_list, g2_vals, lw=1.5)

ax.set_yscale('log')
ax.set_ylim(0, 1e10)
ax.set_xlim(0.0, -5.0)   
ax.set_xlabel(r'$\Delta/\omega_b$')
ax.set_ylabel(r'$g^{(2)}(0)$')
ax.grid(True, which='both', ls=':', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import qutip as qt
import matplotlib.pyplot as plt

omega_b = 1.0
Omega_over_ob = 0.003     
kappa_over_ob = 0.002     
gamma_over_ob = 0.0002    
gphi_over_ob  = 0.0004 

# --- rejillas: Δ de -6→0 (lineal) y λ/ω_b en escala log (10^{-2}→10^{0}) ---
Delta_list   = np.linspace(-6.0, 0.0, 361)
lambda_list  = np.logspace(-2, 0, 61)   # 0.01 ... 1.0 (eje Y logarítmico)

# --- convergencia: con λ hasta 1.0, sube el corte ---
Ncut = 40      # 40–60 recomendado si barres λ/ω_b hasta 1
nbar_floor = 1e-9

# Operadores (reconstruir por el nuevo Ncut)
b   = qt.destroy(Ncut)
I_b = qt.qeye(Ncut)
I_q = qt.qeye(2)
sm  = qt.sigmam(); sp = sm.dag()

proj_e = qt.tensor(sp*sm, I_b)
b_sys  = qt.tensor(I_q, b)
sm_sys = qt.tensor(sm, I_b)
sp_sys = sm_sys.dag()
nb_sys = b_sys.dag()*b_sys

# Hamiltoniano fijo y colapsos
H_bare  = omega_b * qt.tensor(I_q, b.dag()*b)
H_drive = Omega_over_ob * (sm_sys + sp_sys)
c_ops = [
    np.sqrt(kappa_over_ob) * b_sys,
    np.sqrt(gamma_over_ob) * sm_sys,
    np.sqrt(gphi_over_ob)  * proj_e,
]

def g2_of_rho(rho, floor=nbar_floor):
    nbar = qt.expect(nb_sys, rho)
    if nbar <= floor:
        return np.nan
    num = qt.expect(b_sys.dag()*b_sys.dag()*b_sys*b_sys, rho)
    return float(np.real(num / (nbar**2)))

def rho_ss_robusto(H):
    # iterativo primero; si falla, SVD; si falla, directo
    try:
        return qt.steadystate(H, c_ops, method='iterative', tol=1e-12, maxiter=300000, use_precond=True)
    except Exception:
        try:
            return qt.steadystate(H, c_ops, method='svd', tol=1e-12)
        except Exception:
            return qt.steadystate(H, c_ops, method='direct')

# --- calcula el mapa ---
g2_map = np.full((len(lambda_list), len(Delta_list)), np.nan, dtype=float)
for iL, lam_over_ob in enumerate(lambda_list):
    H_int = lam_over_ob * qt.tensor(sp*sm, (b + b.dag()))
    for iD, Delta_over_ob in enumerate(Delta_list):
        H = (Delta_over_ob * proj_e) + H_bare + H_int + H_drive
        rho = rho_ss_robusto(H)
        g2_map[iL, iD] = g2_of_rho(rho)

# --- escala de color como en el panel: log10 g^{(2)} entre 0 y 9 ---
log_g2 = np.log10(np.clip(g2_map, 1.0, 1e9))

# --- ploteo con eje Y logarítmico (usar pcolormesh) ---
X, Y = np.meshgrid(Delta_list, lambda_list)
fig, ax = plt.subplots(figsize=(8.6, 5.2))
pc = ax.pcolormesh(X, Y, log_g2, shading='auto', vmin=0.0, vmax=9.0, cmap='summer')

# líneas guía Δ = -n ω_b (verticales punteadas)
for n in range(1, 7):
    ax.axvline(-n, ls='--', lw=0.8, color='k', alpha=0.6)

cbar = plt.colorbar(pc, ax=ax, pad=0.02)
cbar.set_label(r'$\log_{10}\, g^{(2)}(0)$')
cbar.set_ticks(range(0,10))
cbar.set_ticklabels([str(t) for t in range(0,10)])

ax.set_xlabel(r'$\Delta/\omega_b$')        # -6 → 0 (izq→der)
ax.set_ylabel(r'$\lambda/\omega_b$')       # eje Y log
ax.set_yscale('log')
ax.set_xlim(-6.0, 0.0)
ax.set_ylim(1e-2, 1e0)
ax.set_title(r'Fig. 3(b) — mapa $\log_{10} g^{(2)}(0)$ en $(\Delta/\omega_b,\ \lambda/\omega_b)$, $\Omega/\omega_b=0.003$')
plt.tight_layout()
plt.show()
```

```python
# === Fig. 4(d) — Código con correcciones mínimas sugeridas ===
# Mantengo tu esquema (mesolve y selección por máximo de flujo),
# añado 1) imports del colorbar, 2) tolerancias numéricas,
# 3) corrección de dims en el permutador y 4) evito colisión de nombres.

import numpy as np
import qutip as qt
from matplotlib import pyplot as plt
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable

# --- parámetros (los tuyos) ---
Delta_over_ob = -2.0
lam_over_ob   = 0.1
Omega_over_ob = 0.2
kappa_over_ob = 0.002
gamma_over_ob = 0.0002
gphi_over_ob  = 0.0004

# --- base del sistema ---
N  = 3  # (opcional: N=4 si quieres acolchar un nivel)
a  = qt.tensor(qt.destroy(N), qt.qeye(2))         # oscilador ⊗ I_qubit
sm = qt.tensor(qt.qeye(N), qt.sigmam())           # I_oscil ⊗ σ-
sp = sm.dag()
sx = qt.tensor(qt.qeye(N), qt.sigmax())           # I_oscil ⊗ (σ+ + σ-)
proj_e = sp * sm                                  # |c><c|

# --- Hamiltoniano (ℏ=1, marco rotante) ---
H = (
    Delta_over_ob * proj_e
    + (a.dag() * a)
    + lam_over_ob * proj_e * (a + a.dag())
    + Omega_over_ob * sx
)

# --- colapsos (κ a, γ σ-, γφ |c><c|) ---
c_ops = [
    np.sqrt(kappa_over_ob) * a,
    np.sqrt(gamma_over_ob) * sm,
    np.sqrt(gphi_over_ob)  * proj_e
]

# --- estado inicial (tu elección práctica para “ver” población en |2>) ---
ket_v, ket_c = qt.basis(2, 0), qt.basis(2, 1)
psi0 = (qt.tensor(qt.basis(N, 0), ket_v) + qt.tensor(qt.basis(N, 2), ket_c)).unit()

# --- dinámica y selección de t* (máximo flujo de salida fonónica) ---
opts = qt.Options(nsteps=1_000_000, atol=1e-10, rtol=1e-10)

tmax  = 8.0 / max(kappa_over_ob, 1e-12)
tlist = np.linspace(0.0, tmax, 2001)

nb_sys   = a.dag() * a
res_flux = qt.mesolve(H, psi0, tlist, c_ops, [nb_sys], options=opts, progress_bar=None)
flux_t   = kappa_over_ob * np.real(res_flux.expect[0])

idx_star = int(np.nanargmax(flux_t))
t_star   = float(tlist[idx_star])

# --- estado en t* ---
rho_tstar = qt.mesolve(H, psi0, [t_star], c_ops, [], options=opts, progress_bar=None).states[-1]

# --- (opcional) permutación de la base — no requerida con estos labels ---
USE_SWAP = False

def permutation_operator_from_old_indices(old_indices):
    """Dado el mapeo new_index -> old_index, crea P tal que |e_new> = P |e_old>."""
    dim = len(old_indices)
    P_mat = np.zeros((dim, dim), dtype=complex)
    for new_i, old_i in enumerate(old_indices):
        P_mat[new_i, old_i] = 1.0
    # dims coherentes con (oscilador ⊗ qubit)
    return qt.Qobj(P_mat, dims=[[N, 2], [N, 2]])

if USE_SWAP:
    old_indices = [1, 0, 3, 2, 5, 4]
    P = permutation_operator_from_old_indices(old_indices)
    rho_plot = P * rho_tstar * P.dag()
else:
    rho_plot = rho_tstar

# --- etiquetas en el orden natural de tensor(basis(N,n), basis(2,s)) ---
kets_labels = [r"$|0,v\rangle$", r"$|0,c\rangle$",
               r"$|1,v\rangle$", r"$|1,c\rangle$",
               r"$|2,v\rangle$", r"$|2,c\rangle$"]

# --- gráfica tipo matrix_histogram (módulo de la matriz densidad) ---
try:
    fig, ax = qt.matrix_histogram(rho_plot, kets_labels, kets_labels,
                                  limits=[0, 1], colorbar=False)
    add_manual_cbar = True
except TypeError:
    fig, ax = qt.matrix_histogram(rho_plot, kets_labels, kets_labels, limits=[0, 1])
    add_manual_cbar = False

ax.view_init(azim=-55, elev=45)
ax.set_zlim(0, 1)

if add_manual_cbar:
    norm = Normalize(vmin=0.0, vmax=1.0)
    smap = ScalarMappable(norm=norm, cmap=plt.get_cmap())
    smap.set_array([])
    cbar = plt.colorbar(smap, ax=ax, pad=0.06, shrink=0.7)
    cbar.set_ticks([0, 0.25, 0.5, 0.75, 1.0])
    cbar.set_label(r"$|\rho_{ij}|$")

plt.tight_layout()
plt.show()
```

```python
# === Fig. 4(e) — Matriz densidad ENTRE los dos saltos del canal a (n=2) ===
import numpy as np
import qutip as qt
from matplotlib import pyplot as plt
from matplotlib.colors import Normalize
from matplotlib.cm import ScalarMappable

# --- parámetros (mismos de Fig. 4) ---
Delta_over_ob = -2.0       # si quieres afinar, microbarrido alrededor de Δ2(Ω)
lam_over_ob   = 0.1
Omega_over_ob = 0.2
kappa_over_ob = 0.002
gamma_over_ob = 0.0002
gphi_over_ob  = 0.0004

# --- base del sistema ---
N  = 3  # suficiente para |0>,|1>,|2>
a  = qt.tensor(qt.destroy(N), qt.qeye(2))
sm = qt.tensor(qt.qeye(N), qt.sigmam())
sp = sm.dag()
sx = qt.tensor(qt.qeye(N), qt.sigmax())
proj_e = sp*sm  # |c><c|

# --- Hamiltoniano (ℏ=1) ---
H = ( Delta_over_ob*proj_e
    + a.dag()*a
    + lam_over_ob*proj_e*(a + a.dag())
    + Omega_over_ob*sx )

# --- colapsos (IMPORTANTE: a va PRIMERO para que su índice sea 0) ---
c_ops = [
    np.sqrt(kappa_over_ob) * a,
    np.sqrt(gamma_over_ob) * sm,
    np.sqrt(gphi_over_ob)  * proj_e,
]

# --- estado inicial (cómodo para ver población en |2>) ---
ket_v, ket_c = qt.basis(2,0), qt.basis(2,1)
psi0 = (qt.tensor(qt.basis(N,0), ket_v) + qt.tensor(qt.basis(N,2), ket_c)).unit()

# --- tiempo de simulación (varias vidas de cavidad para capturar una cascada) ---
opts = qt.Options(nsteps=1_000_000, atol=1e-10, rtol=1e-10)
tmax  = 8.0 / max(kappa_over_ob, 1e-12)
tlist = np.linspace(0.0, tmax, 2001)

# --- 1) Corre UNA trayectoria para detectar los dos primeros saltos del canal a ---
seed = 2025  # fija la semilla para reproducibilidad
res_traj = qt.mcsolve(H, psi0, tlist, c_ops, e_ops=[], ntraj=1,
                      seeds=[seed], options=opts, progress_bar=None)

# Tiempos de colapso y qué canal colapsó
col_t = res_traj.col_times
col_w = res_traj.col_which

# Filtra sólo los saltos del canal a (índice 0 en c_ops)
a_clicks = [t for t, w in zip(col_t, col_w) if w == 0]

if len(a_clicks) < 2:
    raise RuntimeError("No se detectaron dos saltos de 'a' en esta trayectoria. "
                       "Aumenta tmax o cambia la semilla.")

t1, t2 = a_clicks[0], a_clicks[1]
t_e = 0.5*(t1 + t2)  # instante ENTRE los dos saltos (Fig. 4e)

# --- 2) Repite la MISMA trayectoria y evalúa el estado exactamente en t_e ---
#    (misma semilla garantiza el mismo patrón de saltos)
res_state = qt.mcsolve(H, psi0, [t_e], c_ops, e_ops=[], ntraj=1,
                       seeds=[seed], options=opts, progress_bar=None)
rho_te = res_state.states[-1]  # densidad en t_e

# --- 3) Grafica |rho_ij(t_e)| con el orden de base natural: |0,v>,|0,c>,|1,v>,... ---
kets_labels = [r"$|0,v\rangle$", r"$|0,c\rangle$",
               r"$|1,v\rangle$", r"$|1,c\rangle$",
               r"$|2,v\rangle$", r"$|2,c\rangle$"]

try:
    fig, ax = qt.matrix_histogram(rho_te, kets_labels, kets_labels,
                                  limits=[0, 1], colorbar=False)
    add_manual_cbar = True
except TypeError:
    fig, ax = qt.matrix_histogram(rho_te, kets_labels, kets_labels, limits=[0, 1])
    add_manual_cbar = False

ax.view_init(azim=-55, elev=45)
ax.set_zlim(0, 1)

if add_manual_cbar:
    norm = Normalize(vmin=0.0, vmax=1.0)
    smap = ScalarMappable(norm=norm, cmap=plt.get_cmap())
    smap.set_array([])
    cbar = plt.colorbar(smap, ax=ax, pad=0.06, shrink=0.7)
    cbar.set_ticks([0, 0.25, 0.5, 0.75, 1.0])
    cbar.set_label(r"$|\rho_{ij}|$")

plt.title(r"Fig. 4(e): estado entre dos emisiones de fonón (después del 1.º salto de $a$)")
plt.tight_layout()
plt.show()
```
