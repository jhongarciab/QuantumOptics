```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
Ia = qeye(3)
Ib = qeye(3)
Is = qeye(2)

a = destroy(3)
b = destroy(3)
s = sigmap()

A = tensor(a, Ib, Is)

def Corre(D,g,J,Ea,Eb,theta):
  H = (D * tensor(a.dag() * a, Ib, Is)
      + D * tensor(Ia, b.dag() * b, Is)
      + D * tensor(Ia, Ib, s.dag() * s)
      + g * (tensor(Ia, b, s.dag()) + tensor(Ia, b.dag(), s))
      + J * (tensor(a.dag(), b, Is) + tensor(a, b.dag(), Is))
      + Ea * tensor((a.dag() * np.exp(-1j*theta) + a * np.exp(1j*theta)), Ib, Is)
      + Eb * tensor(Ia, (b.dag() + b), Is))

  ops = [np.sqrt(kappa)*tensor(a,Ib,Is),
        np.sqrt(kappa)*tensor(Ia,b,Is),
        np.sqrt(gamma)*tensor(Ia,Ib,s)]

  rho_ss = steadystate(H, ops)
  n_ss = expect(tensor(a.dag() * a, Ib, Is), rho_ss)

  G2 = expect(tensor(a.dag() * a.dag() * a * a, Ib, Is), rho_ss)
  g2 = G2 / (n_ss ** 2)
  return (n_ss,g2)
```

#### Figura 2

```python
Ea = 2*np.pi
Eb = 0
theta = 0

kappa = 32*np.pi
g = kappa
gamma = 2*np.pi

D = np.linspace(-2*kappa, 2*kappa, 100)
index = [0.5, 1, 2, 3]
J = [kappa*i for i in index]

L1 = []
L2 = []
for j in J:
  Correl = []
  Foton = []
  for d in D:
    val = Corre(d,g,j,Ea,Eb,theta)
    Correl.append(val[1])
    Foton.append(val[0])
  L1.append(Correl)
  L2.append(Foton)
```

```python
for i, j in enumerate(L1):
  plt.plot(D, j, label='J='+str(index[i])+r'$\kappa$', lw=2)
plt.xlabel(r'$\Delta$')
plt.xticks([-2*kappa, -kappa, 0, kappa, 2*kappa], [r'-2$\kappa$', r'-$\kappa$', '0', r'$\kappa$', r'2$\kappa$'])
plt.ylabel(r'$g^{(2)}$')
plt.legend()
plt.yscale('log')
plt.show()
```

```python
for i, j in enumerate(L2):
  plt.plot(D, j, label='J='+str(index[i])+r'$\kappa$', lw=2)
plt.xlabel(r'$\Delta$')
plt.xticks([-2*kappa, -kappa, 0, kappa, 2*kappa], [r'-2$\kappa$', r'-$\kappa$', '0', r'$\kappa$', r'2$\kappa$'])
plt.ylabel(r'$N$')
plt.legend()
plt.show()
```

```python
Ia = qeye(6)
Ib = qeye(6)
Is = qeye(2)

a = destroy(6)
b = destroy(6)
s = sigmap()

A = tensor(a, Ib, Is)

def Corre(D,g,J,Ea,Eb,theta):
  H = (D * tensor(a.dag() * a, Ib, Is)
      + D * tensor(Ia, b.dag() * b, Is)
      + D * tensor(Ia, Ib, s.dag() * s)
      + g * (tensor(Ia, b, s.dag()) + tensor(Ia, b.dag(), s))
      + J * (tensor(a.dag(), b, Is) + tensor(a, b.dag(), Is))
      + Ea * tensor((a.dag() * np.exp(-1j*theta) + a * np.exp(1j*theta)), Ib, Is)
      + Eb * tensor(Ia, (b.dag() + b), Is))

  ops = [np.sqrt(kappa)*tensor(a,Ib,Is),
        np.sqrt(kappa)*tensor(Ia,b,Is),
        np.sqrt(gamma)*tensor(Ia,Ib,s)]

  rho_ss = steadystate(H, ops)
  n_ss = expect(tensor(a.dag() * a, Ib, Is), rho_ss)

  G2 = expect(tensor(a.dag() * a.dag() * a * a, Ib, Is), rho_ss)
  g2 = G2 / (n_ss ** 2)
  return (n_ss,g2)

Eb = 0
theta = 0
kappa = 16*np.pi
gamma = 1*np.pi
D = 0
g = 1.1*kappa
J = 3*kappa

Ea = np.linspace(0.001*kappa, 3*kappa, 100)

A2 = []
A3 = []
for i in Ea:
  val = Corre(D,g,J,i,Eb,theta)
  A2.append(val[1])
  A3.append(val[0])

fig, ax1 = plt.subplots()
ax1.set_xlabel(r'$E/\kappa$')
ax1.set_ylabel(r'$g^{(2)}(0)$', color='black')
ax1.plot(Ea, A2, 'k-', label=r'$g^{(2)}(0)$')
ax1.tick_params(axis='y', labelcolor='black')
ax1.set_xticks([0, kappa, 2*kappa, 3*kappa], ['0', r'$\kappa$', r'2$\kappa$', r'3$\kappa$'])

ax2 = ax1.twinx()
ax2.set_ylabel(r'$N$', color='darkorange')
ax2.plot(Ea, A3, color='darkorange', label=r'$N$')
ax2.tick_params(axis='y', labelcolor='darkorange')
ax2.set_yticks([0, 0.05, 0.10, 0.15, 0.20], ['0', '0.05', '0.10', '0.15', '0.20'])

ax2.set_ylim(0, 0.20)
plt.show()
```
